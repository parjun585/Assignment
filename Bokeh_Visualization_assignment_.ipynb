# 2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Glyphs are visual markers or shapes used to represent data points on a plot. They can be simple shapes like circles, squares, or lines, or more complex shapes like wedges, patches, or even custom shapes. Glyphs are added to a Bokeh plot using glyph methods such as p.circle(), p.square(), p.line(), etc.

```python
# random data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

output_file("glyphs.html")
p = figure(title="Glyphs Example", x_axis_label='X-axis', y_axis_label='Y-axis')
p.circle(x, y, size=35, color="navy", alpha=0.5)  # Circle glyphs
p.square(x, y, size=12, color="red", alpha=0.8)   # Square glyphs
p.line(x, y, line_width=2, color="green")         # Line glyphs
show(p)
```

#3 How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

By adding the code, mentioning the title, x_axis_label , the y_axis_label, legend_label a name we can customize the plot. for example using the same plot above adding a legend_label to that


```python
# random data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

output_file("glyphs.html")
p = figure(title="Glyphs Example", x_axis_label='X-axis', y_axis_label='Y-axis')
p.circle(x, y, size=35, color="navy", alpha=0.5,legend_label="Circle Glyph")  # Circle glyphs
p.square(x, y, size=12, color="red", alpha=0.8, legend_label="Line Glyph")   # Square glyphs
p.line(x, y, line_width=2, color="green")         # Line glyphs
show(p)
```

#Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

```python
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import file_html
from bokeh.resources import CDN

app = Flask(__name__)

@app.route('/')
def index():
    # Create a Bokeh plot
    p = figure(title="Flask Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis')
    p.circle([1, 2, 3], [4, 5, 6], size=10, color="navy", alpha=0.5)

    # Export the plot as HTML
    plot_html = file_html(p, CDN, "Flask Bokeh Plot")

    # Render the HTML template with the plot
    return render_template('index.html', plot_html=plot_html)

if __name__ == '__main__':
    app.run(debug=True)
```
